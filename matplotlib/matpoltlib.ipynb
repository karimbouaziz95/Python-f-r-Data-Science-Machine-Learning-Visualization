```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
plt.plot(x,y, "r")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.title("Diagramm Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(y,x,"g*-")
```

```python
af = plt.figure()
axes = af.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y,"b")
axes.set_xlabel("xaxis")
axes.set_ylabel("yaxis")
axes.set_title("Uberschrift")
```

```python
af = plt.figure()
axes = af.add_axes([0.5, 0.1, 0.5, 0.8])
axes.plot(x,y,"b")
axes.set_xlabel("xaxis")
axes.set_ylabel("yaxis")
axes.set_title("Uberschrift")
```

```python
af = plt.figure()

axes1 = af.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = af.add_axes([0.2, 0.2, 0.4, 0.3])

axes1.plot(x,y,"b")
axes1.set_xlabel("xaxis")
axes1.set_ylabel("yaxis")
axes1.set_title("das grosse diagramm")

axes2.plot(y,x,"r")
axes2.set_xlabel("xaxis")
axes2.set_ylabel("yaxis")
axes2.set_title("das kleine diagramm")

```

```python
diag, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("Title")

plt.tight_layout()
```
